# Embeding:

Terms:


*   corpus- all words
*   vocabulory-unique words


*   document- each observation/record





```python
d={'good teaching':1,
   'bad teaching':0,
   'amazing teacher':1,
   'wrost teacher':0,
   'excellent teacher':1}

w=[]
for i in d:
  w +=(i.split())
print('corpus:',len(w))
print('vocabulory:',len(set(w)))
print('document:',len(d))
```

```python
print(w)
v={}
print(set(w))
for i in set(w):
  v[i]=w.count(i)
print(v)
```

```python
l=[]
for i in d:
  l1=[]
  for j in v.keys():
    if j in i:
      l1.append(v[j])
    else:l1.append(0)
  l.append(l1)
print(l)
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

```python
df=pd.DataFrame(l,columns=v.keys())
df
```

```python
dt=np.array(l)
dt
```

```python
dt.shape
```

```python
dt=np.reshape(dt,(dt.shape[0],dt.shape[1],1))
dt
```

```python
dt.shape
```

```python
x=tf.convert_to_tensor(dt,dtype=tf.float32)
x
```

```python
x.shape
```

```python
y=np.array(list(d.values()))
y
```

```python
m = keras.Sequential()
m.add(keras.layers.SimpleRNN(1,activation='sigmoid',input_shape=(7,1)))
m.add(keras.layers.Dense(12,activation='sigmoid'))
m.add(keras.layers.Dense(1,activation='sigmoid'))
m.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
m.summary()
```

```python
m.fit(x,y)
```

```python
p=m.predict(x)
```

```python
voc=5000
```

```python
n=x['title'].copy()
n.head(10)
```

# **Stremmer:**

```python
# stopwatch = unimportant words
nltk.download('stopwords',download_dir=r'C:\Users\Admin\Desktop\Stopwords')
nltk.data.path.append(r'C:\Users\Admin\Desktop\Stopwords')
```

```python
df.shape
```

```python
from nltk.stem.porter import PorterStemmer  # porter stremming algorithm is used ti reduce words to thier base or root from , which helps in capturing the essence of words
ps= PorterStemmer()
cp=[]
for i in n:
  print(i)
  review=re.sub('[^a-zA-Z]',' ',str(i))
  #print(review)
  review=review.lower().split()
  #print(review)

  review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
  #print(review)
  review=' '.join(review)
  #print(review)
  cp.append(review)
```

```python
cp
```

## **One Hot**

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
oh = [one_hot(w,voc)for w in cp]
oh
```

# **Embedding**

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sen=20
em=pad_sequences(oh ,padding='pre',maxlen=20)
print(em)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN
from tensorflow.keras.layers import Dense
```

# **Building RNN**

```python
emb_features=20
n=Sequential()
n.add(Embedding(voc,emb_features,input_length=sen))
n.add(SimpleRNN(10800))
n.add(Dense(1,activation='sigmoid'))
n.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
n.summary()
```

```python
# sen = np.array(oh).shape[1]

# Redefine the model with the correct input length
n = Sequential()
n.add(Embedding(voc, emb_features, input_length=sen))
n.add(SimpleRNN(100))
n.add(Dense(1, activation='sigmoid'))
n.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

n.summary()
```

```python
len(em),y.shape
```

```python
x=np.array(em)
y=np.array(y)
```

```python
x.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
xtn,xts,ytn,yts=train_test_split(x,y,test_size=0.2,random_state=8)
```

```python
n.fit(xtn,ytn,epochs=20)
```

```python
pred=n.predict(xts)
print(len(pred))
pred
```
